```python
import pandas as pd
import json
```

```python
with open("./dataset/datas/wp_posts.json") as f:
    tmp = json.loads(f.read())
```

```python
df = pd.DataFrame(tmp[2]['data'])
```

```python
df
```

```python
# 去除数据中的html
import re
from bs4 import BeautifulSoup
def clean_post_content(x):
    soup = BeautifulSoup(x,'html.parser')
    res = soup.get_text()
    return re.sub(r"\r|\n|\t","",res)

df["post_content"] = df["post_content"].map(clean_post_content)
```

```python
df
```

### 使用jieba进行中文分词

```python
import jieba
```

```python
df["post_title"] +","+df["post_content"]
```

```python
def do_cut_words(param_df):
    # 标题加上关键词，是整个待分词的句子
    sentence = param_df["post_title"] +","+param_df["post_content"]
    
    # 分词
    words = list(jieba.cut(sentence))
    
    # 过滤，小写
    res = []
    for word in words:
        if not word or len(word)==0 or len(word)==1:
            continue
        word= word.lower()
        res.append(word)
    return " ".join(res)

df["words"] = df.apply(do_cut_words,axis=1)
```
