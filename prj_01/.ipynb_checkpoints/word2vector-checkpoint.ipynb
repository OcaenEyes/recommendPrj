```python
# 去除数据中的html
import re
from bs4 import BeautifulSoup
def clean_post_content(x):
    soup = BeautifulSoup(x,'html.parser')
    res = soup.get_text()
    return re.sub(r"\r|\n|\t","",res)

df["post_content"] = df["post_content"].map(clean_post_content)
```

```python
df
```

### 使用jieba进行中文分词

```python
import jieba
```

```python
df["post_title"] +","+df["post_content"]
```

```python
def do_cut_words(param_df):
    # 标题加上关键词，是整个待分词的句子
    sentence = param_df["post_title"] +","+param_df["post_content"]
    
    # 分词
    words = list(jieba.cut(sentence))
    
    # 过滤，小写
    res = []
    for word in words:
        if not word or len(word)==0 or len(word)==1:
            continue
        word= word.lower()
        res.append(word)
    return res

df["words"] = df.apply(do_cut_words,axis=1)
```

```python
#对每个句子的所有词向量取均值，来生成一个句子的vector
def build_sentence_vetor(sentence):
    sen_vec = np.zeros(100)
    count =0
    for word in sentence:
        try:
            sen_vec += w2v.wv[word]
            count +=1
        except KeyError:
            continue
    # if count!=0:
    #     sen_vec /=count
    
    return sen_vec
```

```python
df["word2V"]=df["words"].map(build_sentence_vetor)
```
